```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from  keras.layers import Dense
from  keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from google.colab import files
```

```python
file = files.upload()
np.random.seed(42)
```

```python
dataframe = pd.read_csv("/content/sonar.csv",header=None)
dataframe.head()
```

```python
dataset = dataframe.values
x = dataset[:,0:60].astype(float)
y = dataset[:,60]
```

# Converting String to Binary

```python
encoder = LabelEncoder()
encoder.fit(y)
encoded_y = encoder.transform(y)
```

# Neural Architecture


```python
def create_baseline():
  model = Sequential()
  model.add(Dense(60,input_dim=60,kernel_initializer='normal',activation='relu'))
  model.add(Dense(1,kernel_initializer='normal',activation='sigmoid'))
  model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
  return model
```

# Evaluate model with standardized dataset

```python
estimator = KerasClassifier(build_fn=create_baseline, epochs=100, batch_size=5, verbose=0)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
results = cross_val_score(estimator, x, encoded_y, cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" %(results.mean()*100, results.std()*100))
```

Baseline: 80.31% (11.17%)
This is an excellent score without doing any hard work.

```python
estimators = []
estimators.append(('Standardize',StandardScaler()))
estimators.append(('mlp',KerasClassifier(build_fn=create_baseline,epochs=100,batch_size=5,verbose=0))
)
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=10,shuffle=True,random_state=42)
results = cross_val_score(pipeline,x,encoded_y,cv=kfold)
print("Standardized: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

# Feature Extraction - Reducing the neurons in the model architecture

```python
def create_baseline():
  model = Sequential()
  model.add(Dense(30,input_dim=60,kernel_initializer='normal',activation='relu'))
  model.add(Dense(1,kernel_initializer='normal',activation='sigmoid'))
  model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
  return model
estimators = []
estimators.append(('Standardize',StandardScaler()))
estimators.append(('mlp',KerasClassifier(build_fn=create_baseline,epochs=100,batch_size=5,verbose=0))
)
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=10,shuffle=True,random_state=42)
results = cross_val_score(pipeline,x,encoded_y,cv=kfold)
print("Standardized: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

This is a great result because we are
doing slightly better with a network half the size, which in turn takes half the time to train.

# Use a Larger Network Topology

```python
def create_baseline():
  model = Sequential()
  model.add(Dense(60,input_dim=60,kernel_initializer='normal',activation='relu'))
  model.add(Dense(30,kernel_initializer='normal',activation='relu'))
  model.add(Dense(1,kernel_initializer='normal',activation='sigmoid'))
  model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
  return model
estimators = []
estimators.append(('Standardize',StandardScaler()))
estimators.append(('mlp',KerasClassifier(build_fn=create_baseline,epochs=100,batch_size=5,verbose=0))
)
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=10,shuffle=True,random_state=42)
results = cross_val_score(pipeline,x,encoded_y,cv=kfold)
print("Standardized: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```
